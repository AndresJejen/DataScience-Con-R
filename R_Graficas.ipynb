## Gráficas
Las gráficas en R funcionan a partir de un concepto llamado capas (layers).
Cada gráfica diferente es denominada como capa, en la gráfica de abajo se obtiene una gráfica típica de scatter con la capa $geom_point()$. Cada capa se reconoce dentro de la libreria ggplot del paquete tidyverse como objeto geometrico (geom_).

Los objetos geoméricos más populares y con los cuales te puedes ir familiarizando son los siguientes.

1. geom_point()
2. geom_smooth()
3. geom_bar()
4. geom_abline()

Para encontrar documentación rapida sobre cada geom ejecuta los siguientes comandos.

?geom_point

?geom_smooth

?geom_bar

?geom_abline

```r
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy))
```

```r
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy, color = class))
```
